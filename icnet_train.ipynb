```python
def sub_loss(output, label, size):
    label = tf.image.resize(label, (label.shape[1] // size, label.shape[2] // size))
    output = tf.dtypes.cast(output,tf.float32)
    label = tf.dtypes.cast(label,tf.float32)
    loss = tf.keras.losses.sparse_categorical_crossentropy(label, output, from_logits= True)
    loss = tf.reduce_mean(loss)
    return loss
```

```python
@tf.function
def train_step(images, label):
    label_shape = label.shape
    with tf.GradientTape() as tape:
        pred_img = model(images, training=True)
        
        loss_sub4 =  sub_loss(pred_img[2], label, 4)
        loss_sub24 =  sub_loss(pred_img[1], label, 8)
        loss_sub124 =  sub_loss(pred_img[0], label, 16)
        
        loss = loss_sub4 + loss_sub24 +loss_sub124

    gradients_of_model = tape.gradient(loss, model.trainable_variables)
    optimizer.apply_gradients(zip(gradients_of_model, model.trainable_variables))
    
    
    
    train_mIoU(label, pred_mask)

    train_loss(loss)
    train_accuracy1(tf.image.resize(label, (label_shape[1] // 4, label_shape[2] // 4)), pred_img[2])
    train_accuracy2(tf.image.resize(label, (label_shape[1] // 8, label_shape[2] // 8)), pred_img[1])
    train_accuracy3(tf.image.resize(label, (label_shape[1] // 16, label_shape[2] // 16)), pred_img[0])


@tf.function
def test_step(images, label):
    label_shape = label.shape
    pred_img = model(images)
        
    loss_sub4 =  sub_loss(pred_img[2], label, 4)
    loss_sub24 =  sub_loss(pred_img[1], label, 8)
    loss_sub124 =  sub_loss(pred_img[0], label, 16)
    
    loss = loss_sub4 + loss_sub24 +loss_sub124
    
    test_loss(loss)
    test_accuracy1(tf.image.resize(label, (label_shape[1] // 4, label_shape[2] // 4)), pred_img[2])
    test_accuracy2(tf.image.resize(label, (label_shape[1] // 8, label_shape[2] // 8)), pred_img[1])
    test_accuracy3(tf.image.resize(label, (label_shape[1] // 16, label_shape[2] // 16)), pred_img[0])
```
